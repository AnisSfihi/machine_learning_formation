```python
import numpy as np 
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
# On créer un modèle qui contient deux features x1 et x2
# x ne sera alors plus un vecteur mais une matrice de n_samples=100 lignes
# et de 2 colonnes 
x, y =make_regression(n_samples=100, n_features=2, noise=10)
plt.scatter(x[:,0], y)
```

```python
print(x.shape)
y=y.reshape(y.shape[0], 1)
print(y.shape)
```

```python
# x1 et x2 et une colonne de 1 comme résultat
# Si on écrit la commande 
# X = np.hstack((x, np.ones(x.shape))) 
# On aura deux colonnes x1 et x2 car x est une matrice maintenant
# et on aura deux colonnes de 1 pas une car np.ones(x.shape) prendra
# les dimensions de la nouvelle matrice x qui veut dire deux colonnes
# La on écrit np.ones((x.shape[O], 1)) pour avoir 100 lignes et une 
# seule colonne
X=np.hstack((x, np.ones((x.shape[0], 1))))
print(X.shape)
print(X[:10]) 
```

```python
# On ajoute un paramètre à teta car nous avons 3 maintenant a,b et c
teta = np.random.randn(3, 1)
print(teta.shape)
print(teta)
```

```python
def model(X, teta):
    return X.dot(teta)
```

```python
plt.scatter(x[:,0], y)
plt.plot(x[:,0], model(X, teta), c='r')
plt.scatter(x[:,0], model(X,teta), c='r')
```

```python
def cost_function(X, y, teta):
    m=len(y)
    return 1/(2*m) * np.sum((model(X, teta)-y)**2)
```

```python
def grad(X, y, teta):
    m=len(y)
    return 1/m * X.T.dot(model(X, teta)-y)
```

```python
def gradient_descent(X, y, teta, alpha, n_iterations):
    cost_history = np.zeros(n_iterations)
    for i in range(0, n_iterations):
        teta = teta - alpha*grad(X, y, teta)
        cost_history[i] = cost_function(X, y, teta)
    return teta, cost_history
```

```python
teta_final, cost_history = gradient_descent(X, y, teta, alpha=0.01, n_iterations=1000)
teta_final
```

```python
predictions = model(X, teta_final)
# x[:,0] permet d'afficher le modèle par rapport à la variable x1
# x[:,1] permet d'afficher le modèle par rapport à la variable x2
plt.scatter(x[:,1],y)
plt.plot(x[:,1],predictions, c='r')
```

```python
plt.plot(range(1000), cost_history)
# On remarque d'après la courbe de cost_history que la descente de Gradient
# converge effectivement vers la plus petite valeur c'est à dire 
# que la machine est entrain d'apprendre
# Le modèle ne peut mieux apprendre que ça
```

```python
def coef_determination(y, pred):
    u = ((y-pred)**2).sum()
    v = ((y-y.mean())**2).sum()
    return 1- u/v

coef_determination(y, predictions)
```
